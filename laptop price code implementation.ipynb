SVM main algorithm


```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Assuming you have a CSV file with your dataset
file_path = '/content/data.csv'
data = pd.read_csv(file_path)

# Specify the features (X) and the target variable (y)
X = data[['price','resolution_width', 'spec_rating']]  # Adjust feature names accordingly
y = data['display_size']  # Assuming 'price' is the continuous target variable

# Check if there are enough samples for splitting
if len(X) < 2 or len(y) < 2:
    print("Warning: Not enough samples to perform train-test split.")
else:
    # Scale the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=37)

    # Define SVR model with hyperparameter tuning
    param_grid = {
        'C': [0.1, 1, 10],
        'epsilon': [0.01, 0.1, 1],
        'kernel': ['linear', 'rbf']
    }

    svr_model = SVR()
    grid_search = GridSearchCV(estimator=svr_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')

    # Fit the grid search
    grid_search.fit(X_train, y_train)

    # Get the best model from the grid search
    best_svr_model = grid_search.best_estimator_

    # Predict on the test set
    y_pred = best_svr_model.predict(X_test)

    # Calculate accuracy based on a threshold
    threshold = 1.2  # Define your threshold here
    y_pred_binary = np.abs(y_pred - y_test) < threshold
    accuracy = accuracy_score(np.ones_like(y_test), y_pred_binary)  # Since all values are True
    print(f"Accuracy: {accuracy*100:.2f}%")
```

KNN algorithm

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Assuming you have a CSV file with your dataset
file_path = '/content/data.csv'
data = pd.read_csv(file_path)

# Specify the features (X) and the target variable (y)
X = data[['price', 'resolution_width', 'spec_rating']]  # Adjust feature names accordingly
y = data['display_size']  # Assuming 'price' is the continuous target variable

# Check if there are enough samples for splitting
if len(X) < 2 or len(y) < 2:
    print("Warning: Not enough samples to perform train-test split.")
else:
    # Scale the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=37)

    # Define KNeighborsRegressor model with hyperparameter tuning
    param_grid = {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'p': [1, 2]  # 1 for Manhattan distance, 2 for Euclidean distance
    }

    knn_model = KNeighborsRegressor()
    grid_search = GridSearchCV(estimator=knn_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')

    # Fit the grid search
    grid_search.fit(X_train, y_train)

    # Get the best model from the grid search
    best_knn_model = grid_search.best_estimator_

    # Predict on the test set
    y_pred = best_knn_model.predict(X_test)

    # Calculate accuracy based on a threshold
    threshold = 1.00  # Define your threshold here
    y_pred_binary = np.abs(y_pred - y_test) < threshold
    accuracy = accuracy_score(np.ones_like(y_test), y_pred_binary)  # Since all values are True
    print(f"Accuracy: {accuracy*100:.2f}%")
```

RANDOM FOREST algorithm

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
import numpy as np

# Assuming you have a CSV file with your dataset
file_path = '/content/data.csv'
data = pd.read_csv(file_path)

# Specify the features (X) and the target variable (y)
X = data[['price', 'resolution_width', 'spec_rating']]  # Adjust feature names accordingly
y = data['display_size']  # Assuming 'price' is the continuous target variable

# Check if there are enough samples for splitting
if len(X) < 2 or len(y) < 2:
    print("Warning: Not enough samples to perform train-test split.")
else:
    # Scale the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=37)

    # Define RandomForestRegressor model with hyperparameter tuning
    param_grid = {
        'n_estimators': [50, 100, 150],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    rf_model = RandomForestRegressor(random_state=37)
    grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')

    # Fit the grid search
    grid_search.fit(X_train, y_train)

    # Get the best model from the grid search
    best_rf_model = grid_search.best_estimator_

    # Predict on the test set
    y_pred = best_rf_model.predict(X_test)

    # Define a threshold for accuracy-like calculation
    threshold = 1.0  # Define your threshold here
    within_threshold = np.abs(y_pred - y_test) < threshold

    # Calculate accuracy-like metric
    accuracy_like = np.mean(within_threshold)
    print(f"Accuracy-like metric: {accuracy_like * 100:.2f}%")
```

DECISION TREE algorithm

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Assuming you have a CSV file with your dataset
file_path = '/content/data.csv'
data = pd.read_csv(file_path)

# Specify the features (X) and the target variable (y)
X = data[['price', 'resolution_width', 'spec_rating']]  # Adjust feature names accordingly
y = data['display_size']  # Assuming 'price' is the continuous target variable

# Check if there are enough samples for splitting
if len(X) < 2 or len(y) < 2:
    print("Warning: Not enough samples to perform train-test split.")
else:
    # Scale the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    # Define DecisionTreeRegressor model with hyperparameter tuning
    param_grid = {
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    dt_model = DecisionTreeRegressor(random_state=42)
    grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')

    # Fit the grid search
    grid_search.fit(X_train, y_train)

    # Get the best model from the grid search
    best_dt_model = grid_search.best_estimator_

    # Predict on the test set
    y_pred = best_dt_model.predict(X_test)

    # Calculate accuracy based on a threshold
    threshold = 1.0  # Define your threshold here
    y_pred_binary = np.abs(y_pred - y_test) < threshold
    accuracy = accuracy_score(np.ones_like(y_test), y_pred_binary)  # Since all values are True
    print(f"Accuracy: {accuracy*100:.2f}%")
```
